```python
import streamlit as st
import pandas as pd
import io
import matplotlib.pyplot as plt
import seaborn as sns

# Assurez-vous que df (le DataFrame) est dfini avant l'utilisation

# Titre de l'application
st.title('Analyse de la variable "ge"')

# Informations sur le dataset
st.write("## Informations sur le dataset")
buffer = io.StringIO()
df.info(buf=buffer)
s = buffer.getvalue()
st.text(s)

# Description de la variable "ge"
st.write("## Description de l'ge")
st.write(df['age'].describe())

# Distribution de l'ge
st.write("## Distribution de l'ge")
fig, ax = plt.subplots()
sns.histplot(df['age'], bins=20, kde=True, ax=ax)  # Correction: sns.histplot est mieux adapt pour Streamlit
st.pyplot(fig)

# Boxplot de l'ge
st.write("## Boxplot de l'ge")
fig, ax = plt.subplots()
sns.boxplot(x=df['age'], ax=ax)  # Correction: passage direct de la srie df['age']
st.pyplot(fig)

# Analyse des donnes
st.write("## Analyse")
st.write("- L'ge minimum de nos clients est de 18 ans")
st.write("- L'ge maximum de nos clients est de 95 ans")
st.write("- L'ge moyen des clients est de 41 ans")
st.write("- 50% de nos clients ont moins de 39 ans inclu, et 70% de nos clients ont moins de 49 ans inclu")
```
